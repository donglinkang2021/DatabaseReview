```python
import pandas as pd

# 创建表格R
R = pd.DataFrame({
    'A': ['a1', 'a2', 'a3', 'a1', 'a4', 'a2','a1'],
    'B': ['b1', 'b3', 'b4', 'b2', 'b6', 'b2','b2'],
    'C': ['c2', 'c7', 'c6', 'c3', 'c6', 'c3','c1'],
})

# 创建表格S
S = pd.DataFrame({
    'B': ['b1', 'b2', 'b2'],
    'C': ['c2', 'c1', 'c3'],
})
```

```python
R
```

```python
S
```

```python
# 选出R中除去与S属性相同的其余属性，并保持唯一性
R1 = R.drop(['B','C'],axis=1).drop_duplicates()
```

```python
R1
```

```python
# 将R1与S进行笛卡尔积
cartesian_R1xS = R1.assign(key=1).merge(S.assign(key=1), on='key').drop('key', 1)
```

```python
cartesian_R1xS
```

```python
# 求笛卡尔积后的表格与R的差集
R1xS = cartesian_R1xS.drop_duplicates()
R1xS_minus_R = R1xS.append(R).drop_duplicates(keep=False)
R1xS_minus_R
```

```python
tmp = R1xS_minus_R.drop(['B','C'],axis=1).drop_duplicates()
tmp
```

```python
R1.append(tmp).drop_duplicates(keep=False)
```

```python
# 1. 内连接
pd.merge(R, S, how='inner', on=['B', 'C'])
```

```python
# 2. 左连接
pd.merge(R, S, how='left', on=['B', 'C'])
```

```python
# 3. 右连接
pd.merge(R, S, how='right', on=['B', 'C'])
```

```python
# 4. 全连接
pd.merge(R, S, how='outer', on=['B', 'C'])
```
